```python
import pandas as pd
import random
from faker import Faker

fake = Faker()

# Assuming item probability matrices are defined as dictionaries like those shown above
entree_probs = {1: 0.10, 2: 0.15}
beverage_probs = {1: 0.20, 2: 0.25}

def generate_dummy_transactions(num_transactions):
    transactions = []

    for _ in range(num_transactions):
        order_number = fake.unique.random_int(min=1, max=10000)
        order_type = random.choices(['Single', 'Group'], [0.65, 0.35])[0]
        group_size = random.choices([2, 3, 4, 5], [0.4, 0.3, 0.2, 0.1])[0] if order_type == 'Group' else 1
        transaction_date = fake.date_between(start_date='-1y', end_date='today').strftime("%Y-%m-%d")
        transaction_time = fake.time()

        for _ in range(group_size):
            meal_composition = random.choices(['Entree Only', 'Entree & Bev', 'Bev Only'], [0.60, 0.35, 0.05])[0]
            entree_num = random.choices(list(entree_probs.keys()), list(entree_probs.values()))[0] if 'Entree' in meal_composition else None
            beverage_num = random.choices(list(beverage_probs.keys()), list(beverage_probs.values()))[0] if 'Bev' in meal_composition else None

            transactions.append({
                "Order_number": order_number,
                "Transaction_date": transaction_date,
                "Transaction_time": transaction_time,
                "Entree_num": entree_num,
                "Beverage_num": beverage_num
            })

    return pd.DataFrame(transactions)

# Generate the transactions
dummy_transactions_df = generate_dummy_transactions(1000)

# Save the transactions to a CSV file
dummy_transactions_df.to_csv("dummy_transactions.csv", index=False)
```
